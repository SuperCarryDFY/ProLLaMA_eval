## get input

```python
import random 
random.seed(0)

with open("../src/ProLLaMA/superfamilies.txt", 'r') as f:
    content = f.readlines()

random.shuffle(content)
content = [i for i in content if i != "\n"]
lines = content[:500]
```

```python
with open("input.txt", 'w') as f:
    for line in lines:
        for _ in range(5):
            f.write(f"[Generate by superfamily] {line}")
```

## Post-process

After run scripts/infer.sh

```python
## extract text and seqs to jsonl and fasta
res = []
with open("output.txt", 'r') as f:
    for line in f:
        all_splits = []
        splits = line.strip().split("<")
        for split in splits:
            all_splits.extend(split.split(">"))
        try:
            text = all_splits[1]
            seq = all_splits[3]
            res.append((text, seq))
        except:
            continue
            # print(line)

len(res)
```

```python
import json 

with open("seqs.fasta", 'w' ) as f:
    for idx, (text, seq) in enumerate(res):
        f.write(f">idx_{idx}\n{seq}\n")

with open("res.jsonl", "w") as f:
    for idx, (text, seq) in enumerate(res):
        json.dump({"idx": idx, "text": text, "seq": seq}, f)
        f.write("\n")
```
